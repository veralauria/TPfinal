```python
import numpy as np
import matplotlib.pyplot as plt

# Frecuencia [Hz]
f = np.array([10, 100, 1000, 2000, 5000, 10000,
              20000, 30000, 40000, 50000, 60000, 100000])

# Ganancia medida [dB]
db_med = np.array([20.000, 20.000, 20.000, 20.000, 20.000, 20.077,
                   15.857, 11.380, 7.707, 5.864, 3.726, -4.783])

# Ganancia simulada [dB]
db_sim = np.array([20.00, 20.00, 20.00, 20.00, 19.98, 19.93,
                   19.71, 19.44, 19.07, 18.61, 18.11, 15.99])

plt.figure(figsize=(10,6))

plt.semilogx(f, db_med, 'o-', linewidth=2, markersize=6,
             label='Medición')
plt.semilogx(f, db_sim, 's--', linewidth=2, markersize=6,
             label='Simulación')

# Nivel -3 dB
ganancia_bf = np.mean(db_med[:6])
nivel_3db = ganancia_bf - 3

plt.axhline(nivel_3db, linestyle=':', linewidth=1.5,
            label='Nivel -3 dB')

plt.title('Diagrama de Bode - Magnitud')
plt.xlabel('Frecuencia [Hz]')
plt.ylabel('Ganancia [dB]')

plt.xlim(10, 100000)
plt.xticks([10, 100, 1000, 10000, 100000],
           ['10', '100', '1k', '10k', '100k'])

plt.grid(True, which='both', linestyle='--', alpha=0.7)
plt.legend()
plt.tight_layout()

plt.show()
```
